```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("iris.csv")
```

```python
df.columns
```

```python
df.head()
```

```python
df.plot(kind='scatter',x='sepal_length',y='sepal_width')
plt.show()
```

```python
plt.plot(df['sepal_length'],df['sepal_width'],'rx')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.title('EDA')
plt.grid()
plt.show()
```

```python
sns.relplot(x='sepal_length',y='sepal_width',hue='species',data=df,height=7,style='species')
```

```python
sns.set_style('darkgrid')
sns.pairplot(df,hue='species')
plt.show()
```

```python
sns.displot(data=df,hue='species',x="petal_length")
```

```python
g=sns.FacetGrid(df,hue='species',size=5)
g.map(sns.distplot,'petal_length')
plt.show()
```

```python
sns.ecdfplot(data=df,x=df['sepal_length'][df['species']=='setosa'])
```

```python
df['species'].value_counts()
```

```python
iris_virginica=df[df['species']=='virginica']
iris_versicolor=df[df['species']=='versicolor']
iris_setosa=df[df['species']=='setosa']
```

```python
print(np.mean(iris_virginica['petal_length']))
print(np.std(iris_virginica['petal_length']))
print(np.median(iris_virginica['petal_length']))
print(np.percentile(iris_virginica['petal_length'],90))
```

```python
print(np.percentile(iris_setosa['petal_length'],[25,50,75,90]))
```

```python
virgin_median=np.median(iris_virginica['petal_length'])
mad=np.mean(np.abs(iris_virginica['petal_length']-np.median(iris_virginica['petal_length'])))
mad
```

```python
from statsmodels import robust
print(robust.scale.mad(iris_virginica['petal_length']))
```
