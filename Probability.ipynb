```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.utils import resample
import seaborn as sns
```

#QQ Plot

Used to see if a random variable follows a distribution


```python
st_normal=np.random.normal(loc=0,scale=1,size=1000)
st_normal.sort()
st_normal_percentile=[np.percentile(st_normal,i)  for i in range(1,101)]
len(st_normal_percentile)
```

```python
to_check=np.random.normal(loc=25,scale=50,size=100000)
to_check.sort()
to_check_percentile=[np.percentile(to_check,i)  for i in range(1,101)]
len(to_check_percentile)
```

```python
sns.distplot(to_check_percentile,hist=False,color='red')
```

```python
sns.distplot(st_normal_percentile,hist=False)
```

```python
plt.plot(to_check_percentile,st_normal_percentile)
```

#Boxcox Transform

Convert any distribution into Gaussian

```python
power_dist=np.random.power(5,size=1000)
```

```python
sns.distplot(power_dist,hist=False)
```

```python
gauss_power_dist,_=stats.boxcox(power_dist)
gauss_power_dist=list(gauss_power_dist)
```

```python
sns.distplot(gauss_power_dist,hist=False)
```

```python
pareto_dist=np.random.pareto(5,size=1000)
sns.distplot(pareto_dist)
```

```python
gauss_pareto_dist,_=stats.boxcox(pareto_dist)
sns.distplot(gauss_pareto_dist)
```

#Bootstrap CI

Computation method to calculate confidence interval

```python
samp_e=np.random.randint(size=20,low=160,high=170)
print(samp_e)
k=1000
mediann=[]
for i in range(k):
  s=resample(samp_e,n_samples=20)
  mediann.append(np.median(s))
alpha=95
sns.distplot(mediann)
l=np.percentile(mediann,(100-alpha)/2)
u=np.percentile(mediann,alpha+(100-alpha)/2)
print("{}% confidence interval of median is from {} to {}".format(alpha,l,u))
```

#KS Test

```python
x=stats.norm.rvs(size=1000)
sns.kdeplot(np.array(x),bw=0.5)
```

```python
stats.kstest(x,'norm')
```

```python
y=np.random.uniform(0,1,1000)
sns.kdeplot(y,bw=0.1)
```
