#KNN

##Defining Model

```python
# Importing Modules

import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, neighbors
from matplotlib.colors import ListedColormap
```

We try to define a function to colour all the region where KNN will give different outputs 

```python
def knn_model(data,k):
  X=data[:,:2]
  y=data[:,2]

  # grid cell size
  h = 0.02
  cmap_light = ListedColormap(['#FFAAAA',  '#AAAAFF'])
  cmap_bold = ListedColormap(['#FF0000', '#0000FF'])

  # finding upper and lower limit for mesh
  x_min = np.min(X[:,0])-1
  x_max = np.max(X[:,0])+1
  y_min = np.min(X[:,1])-1
  y_max = np.max(X[:,1])+1

  #creating a mesh grid
  xx,yy = np.meshgrid(np.arange(x_min, x_max, h),np.arange(y_min, y_max, h))

  #Defining Model
  knn = neighbors.KNeighborsClassifier(n_neighbors=k)
  knn.fit(X,y)


  # np.c_ : Translates slice objects to concatenation along the second axis.
  # > np.c_[np.array([1,2,3]), np.array([4,5,6])]
  # > array([[1, 4],
  #          [2, 5],
  #          [3, 6]])   (source: np.c_ documentation)
  Z = np.c_[np.ravel(xx),np.ravel(yy)]

  # predicting
  pred = knn.predict(Z)

  # reshaping pred for colouring
  pred = pred.reshape(xx.shape)

  #Plotting the decision boundary
  plt.figure(figsize=(8, 6), dpi=80)
  #pcolourmesh will plot the xx,yy grid with the colour according to pred so it will look like a decision boundary
  plt.pcolormesh(xx, yy, pred, cmap=cmap_light)

  # plotting a scatter plot
  plt.scatter(X[:,0],X[:,1],c=y,cmap=cmap_bold)

  # limits for x and y
  plt.xlim(xx.min(),xx.max())
  plt.ylim(yy.min(),yy.max())

  #setting title
  plt.title("Decision Surface for k = "+str(k))
  plt.show()
```

##For concentric circle 1

```python
print("="*200)
print("="*200)
for i in [1,5,15,30,50]:
  knn_model(concentric1,i)
  print("="*200)
  print("="*200)
```
